```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

file_path = r"a.xlsx"

# Read the Excel file into a pandas DataFrame
data = pd.read_excel(file_path)

# Display the imported data
print(data)
```

```python
data.plot(x='Time', y='Interest rate', kind='line')
```

```python
print(data.columns)
interest_rate = data['Interest rate']

# Plot ACF
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plot_acf(interest_rate, lags=40, ax=plt.gca())
plt.title('ACF of Interest Rate')

# Plot PACF
plt.subplot(1, 2, 2)
plot_pacf(interest_rate, lags=40, ax=plt.gca())
plt.title('PACF of Interest Rate')

plt.tight_layout()
plt.show()
```

```python
interest_rate_squared = interest_rate ** 2

plt.plot(interest_rate_squared)
plt.xlabel('Time')
plt.ylabel('Interest Rate Squared')
plt.title('Plot of Interest Rate Squared')
plt.show()
```

```python


# Plot ACF
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plot_acf(interest_rate_squared, lags=40, ax=plt.gca())
plt.title('ACF of Interest Rate Squared')

# Plot PACF
plt.subplot(1, 2, 2)
plot_pacf(interest_rate_squared, lags=40, ax=plt.gca())
plt.title('PACF of Interest Rate Squared')

plt.tight_layout()
plt.show()
```
